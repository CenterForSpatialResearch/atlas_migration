```python
import numpy as np
import pandas as pd
import geopandas as gpd
import ast # for converting strings ito tuples

# import seaborn as sns
import matplotlib.pyplot as plt
# import matplotlib.patches as mpatches
import plotly.graph_objects as go

import warnings
warnings.filterwarnings('ignore')
```

```python
from datetime import datetime

# Creating a datetime object
d = datetime.now()

# Converting a to string using strftime
datenow = d.strftime('%Y%m%d')
timenow = d.strftime("%H%M%S")

print(datenow)
print(timenow)
```

# Data

```python
star = pd.read_csv("data_gen/trajectories/star_all_distances.csv", index_col=0)

for i, c in enumerate(star.columns):
    print('{}\t{}'.format(i, c))

print(star.location_country)
print(star.city_8)
display(star)
```

```python
pd.DataFrame( star[star['city_8'].eq('Accra') & star['location_type'].eq(100)]['location'].value_counts() ).head(10)
```

# Topmost and Bottommost Origins

## Accra

```python
n=10 # Top 10, Bottom 10
l='Accra'

display( pd.DataFrame( star[star['city_8'].eq(l) & star['location_type'].eq(100)]['location'].value_counts() ).head(n) )
display( pd.DataFrame( star[star['city_8'].eq(l) & star['location_type'].eq(100)]['location'].value_counts() ).tail(n) )
```

```python
l='Accra'

df=star[star['city_8'].eq(l) & star['location_type'].eq(100)][['location', 'lat', 'lon', 'ID']].groupby(['location', 'lat', 'lon']).count().reset_index().sort_values('ID', ascending=False)
display(df)
df.to_csv('{} {} Top Origins'.format(datenow, l))
```
