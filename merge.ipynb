```python
import pandas as pd

# โหลดข้อมูลจากไฟล์
file_path_1 = 'data/level1_migrations_UR.csv'
file_path_2 = 'data/star_all_distances.csv'
data_1 = pd.read_csv(file_path_1)
data_2 = pd.read_csv(file_path_2)

# สร้าง DataFrame ที่ใช้ location_l และ neighborhood_9 จาก data_2
location_neighborhood_df = data_2[['location_l', 'neighborhood_9']].drop_duplicates().reset_index(drop=True)

# ตรวจสอบคู่ที่ไม่ซ้ำเพื่อความถูกต้อง
print("คู่ที่ไม่ซ้ำของ location_l และ neighborhood_9 จาก data_2:")
print(location_neighborhood_df.head(20))  # แสดงคู่แรก 20 คู่เพื่อการตรวจสอบ

# จับคู่ loc1 ใน data_1 กับ location_l ใน data_2 เพื่อหา neighborhood_9 และสร้างคอลัมน์ neighbor
merged_data = pd.merge(data_1, location_neighborhood_df, left_on='loc1', right_on='location_l', how='left')

# เปลี่ยนชื่อคอลัมน์ neighborhood_9 เป็น neighbor
merged_data = merged_data.rename(columns={'neighborhood_9': 'neighbor'})

# ลบคอลัมน์ location_l ที่เกินมา
merged_data = merged_data.drop(columns=['location_l'])

# บันทึกข้อมูลที่อัปเดตลงในไฟล์ CSV ใหม่
output_file_path = 'data/updated_data_with_neighbor.csv'
merged_data.to_csv(output_file_path, index=False)

# แสดงข้อมูลแถวแรกๆ เพื่อการตรวจสอบการจับคู่
print("แถวแรกของข้อมูลที่อัปเดต:")
print(merged_data[['loc1', 'neighbor']].head())
```

```python
import pandas as pd

# Load the CSV files
df_star = pd.read_csv('data/star_all_distances.csv')
df_origin = pd.read_csv('data/origin_migrations_UR.csv')

unique_pairs_star = df_star[['city_8', 'location_l']].drop_duplicates()
unique_pairs_origin = df_origin[['city_8', 'loc2']].drop_duplicates()

print("Unique pairs of city_8 and location_l in star_all_distances.csv:")
print(unique_pairs_star)

print("Unique pairs of city_8 and loc2 in origin_migrations_UR.csv:")
print(unique_pairs_origin)

# คุณสามารถแก้ไขค่าที่ไม่ตรงกันได้ด้วยการสร้าง mapping ที่ถูกต้อง
# สมมุติว่าเราพบว่ามีข้อผิดพลาดและต้องการแก้ไขบางค่าใน df_star
# ตัวอย่างการแก้ไขข้อมูลใน df_star:
# df_star.loc[df_star['neighborhood_9'] == 'Kawangware', 'city_8'] = 'Nairobi'
# df_star.loc[df_star['neighborhood_9'] == 'Kayole', 'city_8'] = 'Nairobi'
# (ทำซ้ำสำหรับคู่ค่าที่ต้องการแก้ไข)

# หลังจากแก้ไขแล้ว สร้าง mapping จาก location_l ไปยัง neighborhood_9
location_to_neighborhood = df_star.set_index('location_l')['neighborhood_9'].to_dict()

# Map the loc2 values to their corresponding neighborhood_9 values
df_origin['neighborhood_9'] = df_origin['loc2'].map(location_to_neighborhood)

# Print the updated dataframe
print(df_origin)

# Save the updated dataframe to a new CSV file
df_origin.to_csv('data/origin_migrations_with_neighborhood.csv', index=False)
```

```python
import pandas as pd

# Load the CSV files
df_star = pd.read_csv('data/star_all_distances.csv')
df_origin = pd.read_csv('data/origin_migrations_UR.csv')

# Specify the column names
column_location_l = 'location_l'
column_loc2 = 'loc2'
column_neighborhood = 'neighborhood_9'

# Create a mapping from location_l to neighborhood_9
location_to_neighborhood = df_star.set_index(column_location_l)[column_neighborhood].to_dict()

# Map the loc2 values to their corresponding neighborhood_9 values
df_origin[column_neighborhood] = df_origin[column_loc2].map(location_to_neighborhood)

# Print the updated dataframe
print(df_origin)

# Save the updated dataframe to a new CSV file
df_origin.to_csv('data/new.csv', index=False)
```
